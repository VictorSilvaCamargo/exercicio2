```python
import math
import matplotlib.pyplot as plt

def intensidade(magnitude, intensity_ref=10**-4.5):
    return intensity_ref * 10**(1.5*magnitude - 1.5*2.12)

magnitude_turquia = 7.8
magnitude_indonesia=9.1
magnitude_chile=9.5

intensidade_turquia = intensidade(magnitude_turquia)
intensidade_indonesia = intensidade(magnitude_indonesia)
intensidade_chile = intensidade(magnitude_chile)

print("Intensidade do terremoto da Turquia:", intensidade_turquia)
print("Intensidade do terremoto da Indonesia:", intensidade_indonesia)
print("Intensidade do terremoto do Chile:", intensidade_chile)




magnitudes = [magnitude_turquia, magnitude_indonesia, magnitude_chile]
intensidades = [intensidade_turquia, intensidade_indonesia, intensidade_chile]

plt.plot(magnitudes, intensidades, 'o--')
plt.xlabel('Magnitude')
plt.ylabel('Intensidade')
plt.title('Comparação das intensidades dos terremotos')
plt.show()
```
